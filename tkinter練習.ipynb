```python
import tkinter as tk
window = tk.Tk()
window.title('My Window')
window.geometry('500x300')


        
allname={}
l1=tk.Label(text="首次登入，請輸入帳號密碼")
l1.pack()

v_name=tk.StringVar()
v_password=tk.StringVar()

e1 = tk.Entry(window,text="請輸入帳號",textvariable=v_name)  # 顯示成明文形式
e2 = tk.Entry(window, text="請輸入密碼",textvariable=v_password,show='*')   # 顯示成密文形式

e1.pack()
e2.pack()

def sign():
    w_sign= tk.Toplevel(window)
    w_sign.geometry('300x200')
    w_sign.title('註冊個新帳號吧')
    l2=tk.Label(w_sign,text="密碼不相符")
        
    sn=tk.StringVar()
    sp=tk.StringVar()
    rsp=tk.StringVar()
        
    tk.Entry(w_sign,text="請輸入新帳號：",textvariable=sn).pack()
    tk.Entry(w_sign,text="請輸入新密碼：",textvariable=sp).pack()
    tk.Entry(w_sign,text="請再次輸入密碼：",textvariable=rsp).pack()
        
    if sp.get()!=rsp.get():
        l2.pack()
    else:
        l2.config(text="註冊成功")
        w_sign.destroy()
            
    
def bt():
    name=v_name.get()
    password=v_password.get()
    
    if name in allname:
        if password==allname[name]:
            l1.config(text="登錄成功")
        else:
            l1.config(text="登錄失敗ＱＱ")
            
    else:
        l1.config(text="沒有此帳號ＱＱ")
        w_sign= tk.Toplevel(window)
        w_sign.geometry('300x200')
        w_sign.title('註冊個新帳號吧')
        
        l2=tk.Label(w_sign,text="密碼不相符")

        sn=tk.StringVar()
        sp=tk.StringVar()
        rsp=tk.StringVar()

        tk.Entry(w_sign,text="請輸入新帳號：",textvariable=sn).pack()
        tk.Entry(w_sign,text="請輸入新密碼：",textvariable=sp).pack()
        tk.Entry(w_sign,text="請再次輸入密碼：",textvariable=rsp).pack()

        if sp.get()!=rsp.get():
            l2.pack()
        else:
            l2.config(text="註冊成功")
            w_sign.destroy()
        
        allname[name]=password #創建新的帳戶
    
bt1=tk.Button(window,text="hi",command=bt)
bt1.pack()

window.mainloop()
```

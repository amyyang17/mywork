# 使用Facebook 開發的Prophet進行股價預測
## 程式說明：
單純用歷史資料進行回測，沒有加入籌碼、技術面等分析，透過歷史收盤調整價來預測未來走勢

### 用pandas_datareader 讀入2330台積電的股價，時間從2016開始，並繪出走勢圖

```python
df= pdr.DataReader('2330.tw', 'yahoo',start=(2016))
```

```python
df_2016=df.index
```

```python
df['Adj Close'].plot(figsize=(12, 8))
```

```python
df.head()
```

### import 進要用的模型

```python
from fbprophet import Prophet
```

```python
model=Prophet(daily_seasonality=True)
```

### 把資料類型轉換成prophet的型態

```python
# 因為prophet要求的檔案格式是ds,y因此轉換
new_df= pd.DataFrame(df['Adj Close']).reset_index().rename(columns={'Date':'ds', 'Adj Close':'y'})
```

```python
new_df.head()
```

### 開始進行預測

```python
model.fit(new_df)
future=model.make_future_dataframe(periods=365)#建立一個365天的時間dataframe來放預測的資料
forecast=model.predict(future)
```

#### 說明：黑色為實際走勢，淺藍色為預測範圍，藍色為預測走勢，日子距離越遠不確定性越大

```python
figure=model.plot(forecast)
```

```python
df_close=df['Adj Close']
year= forecast.set_index('ds').join(df_close)#加入實際資料做比對
```

```python
year.head()
```

### 比對實際結果與預測的落差
灰色為上下限，藍色為實際股價，橘色為預測的走勢

```python
fig, ax = plt.subplots(figsize=(12, 8))
ax.plot(year['Adj Close'],label="actual")
ax.plot(year.yhat,label="predict")
ax.plot(year.yhat_upper, color='black',  linestyle=':', alpha=0.5)
ax.plot(year.yhat_lower, color='black',  linestyle=':', alpha=0.5)
ax.set_ylabel('Price')
ax.set_xlabel('Date')
ax.legend(loc='upper left') 
```

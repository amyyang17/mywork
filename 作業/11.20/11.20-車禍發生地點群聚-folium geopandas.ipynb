# 運用folium的markercluster

```python
import folium
from folium.plugins import MarkerCluster
```

```python
map1=folium.Map([23.5,120.5],zoom_start=7)
mc=MarkerCluster()
```

```python
for i in range(0,len(df)):
    marks=folium.Marker([df.緯度[i],df.經度[i]])
    mc.add_child(marks).add_to(map1)
```

```python
map1
```

```python
df.insert(3,"死亡人數",(df["死亡受傷人數"].map(lambda x:x.replace("死亡","").split(";")[0])))
df.insert(4,"受傷人數",(df["死亡受傷人數"].map(lambda x:x.replace("受傷","").split(";")[1])))
```

```python
df.drop(columns="死亡受傷人數",inplace=True)
```

### 將地圖轉換為geopandas 的樣子

```python
gdf= gpd.GeoDataFrame(df, geometry=gpd.points_from_xy(df.經度, df.緯度))
```

# 與jupyter notebook互動

```python
def f(x):
    print(x)
    return x
```

```python
f(76)
```

```python
from ipywidgets import interact
```

```python
interact(f, x=3)
```

```python
interact(f, x=3.)  #只需要加個.即可變浮點數
```

```python
interact(f, x="hi") 
```

```python
interact(f, x=[1,2,3,4])  #下拉式選單(列表)
```

```python
interact(f, x={"台北":1,"台中":2}) #字典功能
```

```python
def myplot (n=1):
    x=np.linspace(-5,5,500)
    y=np.sin(n*x)
    plt.plot(x,y)
```

```python
myplot(3)
```

```python
interact(myplot,n=(1,40))
```

# 作業一 簡易音樂播放器

```python
def music(choose):
    pygame.mixer.init()
    music1 = pygame.mixer.music.load("%s.mp3"%(choose))
    pygame.mixer.music.play()
def music_stop():
    pygame.mixer.music.stop()
```

須將音樂放置在檔案夾內

```python
interact_manual(music,choose=["閻奕格 優雅道別","徐佳瑩 你敢不敢","郁可唯 路過人間 ","孫燕姿 我懷念的"]) 

button= Button(description="停止音樂")
display(button)

def on_button_clicked(a):
    music_stop()

button.on_click(on_button_clicked)
```

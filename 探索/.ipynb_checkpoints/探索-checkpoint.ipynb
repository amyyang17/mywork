```python
import tkinter as tk
#視窗
window=tk.Tk()
window.title("try1")
window.geometry("1000x1000")

#設定圖片
img_1 = tk.PhotoImage(file ='圖片1.png')
img_2 = tk.PhotoImage(file ='圖片2.png')
img_3 = tk.PhotoImage(file ='圖片3.png')
img_21 = tk.PhotoImage(file ='圖片21.png')
img_22 = tk.PhotoImage(file ='圖片22.png')
img_23 = tk.PhotoImage(file ='圖片23.png')
img_31 = tk.PhotoImage(file ='圖片31.png')
img_32 = tk.PhotoImage(file ='圖片32.png')
img_33 = tk.PhotoImage(file ='圖片33.png')
img_41 = tk.PhotoImage(file ='圖片41.png')
img_42 = tk.PhotoImage(file ='圖片42.png')
img_43 = tk.PhotoImage(file ='圖片43.png')

#設定變數
var=tk.IntVar()
var.set(0)
choice=[]
j=0


#歡迎介面
label=tk.Label(window,text="嗨囉，歡迎來到遊戲，底下點擊開始吧",font=('Arial', 20))
label.grid(row=0,column=0,columnspan=5)

#選項(未完成)*1
c1={0:"上街擺攤販售商品維生",1:"遊走各公司只為填飽肚子",2:"到西迪布吉德的一座清真寺附近示威、放火燒輪胎以及攻擊憲政民主聯盟的辦公室"}
c2={0:"聽從船長的指示留在原地",1:"不聽從指示，爬上船頂跳入海中",2:"使用消防水管和窗簾等等工具試圖脫困"}
c3={0:"試圖以橡膠子彈與催淚彈驅散抗議民眾",1:"聽從上級命令嚴厲鎮壓抗議民眾",2:"和其他警察一同站上街頭，抗議自己所受的待遇"}
c4={0:"大力贊成政府派遣軍隊壓制動亂",1:"參與示威民眾搶劫超商、縱火的行動",2:"事不關己，沒有任何作為"}
ca=[c1,c2,c3,c4]

#開始
def bts():
    btstart.destroy()
    label.config(text="2010年茉莉花革命造成突尼西亞各地民不聊生，如果你是當時受迫的老百姓，你會...")
    #最開始設定
    for val,ch in c1.items():
        choice.append(tk.Radiobutton(window,text=ch,variable=var,value=val,command=select,font=('Arial', 20)))
        choice[-1].grid(row=val+3,rowspan=1,column=0,columnspan=5)
    bt1.grid(row=7)
    global l
    l=tk.Label(window,text="你尚未選擇",font=('Arial', 20))
    l.grid(row=2,columnspan=4)

#第一題（已完成）
def bt1():#只會用一次
    global j
    j+=1
    s=var.get()
    
    #加入條件
    if s==0:
        label.config(text="結果：遭到城管誣陷沒有擺攤執照，投訴無門，最後選擇在政府大樓前潑油自焚。")
        global label_img
        label_img = tk.Label(window, image = img_1)
        label_img.grid(row=3,rowspan=4,columnspan=4)
    elif s==1:
        label.config(text="結果：無法忍受飢餓與失業，爬上電塔使自己觸電。")
        label_img = tk.Label(window, image = img_2)
        label_img.grid(row=3,rowspan=4,columnspan=4)
    else:
        label.config(text="結果：遭警察使用催淚瓦斯驅趕不幸失明。")
        label_img = tk.Label(window, image = img_3)
        label_img.grid(row=3,rowspan=4,columnspan=4)
    l.grid_forget()
        
    for i in range(0,3):
        choice[i].destroy()
        
    global btn
    btn=tk.Button(window,text="下一關",command=btnnext,font=('Arial', 20))
    btn.grid(row=7,column=4)
    bt1.grid_forget()
    
#第2題  （已修改）
def bt2():
    global j
    j+=1
    s=var.get()
    if s==0:
        label.config(text="結果：直到事情已無法收拾後，不幸罹難。")
        global label_img
        label_img = tk.Label(window, image = img_21)
        label_img.grid(row=3,rowspan=4,columnspan=4)
    elif s==1:
        label.config(text="結果：幸運獲得附近漁民的救援")
        label_img = tk.Label(window, image = img_22)
        label_img.grid(row=3,rowspan=4,columnspan=4)
    else:
        label.config(text="結果：獲得海警載離事故現場。")
        label_img = tk.Label(window, image = img_23)
        label_img.grid(row=3,rowspan=4,columnspan=4)
    l.grid_forget()
    
    for i in range(0,3):
        choice[i].destroy()
    global btn2
    btn2=tk.Button(window,text="下一關",command=btn3choice,font=('Arial', 20))
    btn2.grid(row=7,column=4)
    bt2.grid_forget()
    
#第3題（未完成）
def bt3():
    global j
    j+=1
    s=var.get()
    if s==0:
        label.config(text="結果：獲得海警載離事故現場")
        global label_img
        label_img = tk.Label(window, image = img_1)
        label_img.grid(row=3,rowspan=4,columnspan=4)
    elif s==1:
        label.config(text="結果：無法忍受飢餓與失業，爬上電塔使自己觸電。")
        label_img = tk.Label(window, image = img_2)
        label_img.grid(row=3,rowspan=4,columnspan=4)
    else:
        label.config(text="結果：遭警察使用催淚瓦斯驅趕不幸失明。")
        label_img = tk.Label(window, image = img_3)
        label_img.grid(row=3,rowspan=4,columnspan=4)
    l.grid_forget()
    
    for i in range(0,3):
        choice[i].destroy()

    bt3.destroy()
    global btn3
    btn3=tk.Button(window,text="下一關",command=btn4choice,font=('Arial', 20))
    btn3.grid(row=7,column=4)
    bt3.grid_forget()
    
#第四題（未完成）
def bt4():
    s=var.get()
    if s==0:
        label.config(text="結果：遭到城管誣陷沒有擺攤執照，投訴無門，最後選擇在政府大樓前潑油自焚。")
        global label_img
        label_img = tk.Label(window, image = img_1)
        label_img.grid(row=3,rowspan=4,columnspan=4)
    elif s==1:
        label.config(text="結果：無法忍受飢餓與失業，爬上電塔使自己觸電。")
        label_img = tk.Label(window, image = img_2)
        label_img.grid(row=3,rowspan=4,columnspan=4)
    else:
        label.config(text="結果：遭警察使用催淚瓦斯驅趕不幸失明。")
        label_img = tk.Label(window, image = img_3)
        label_img.grid(row=3,rowspan=4,columnspan=4)
    l.grid_forget()

    for i in range(0,3):
        choice[i].destroy()
      
    bt4.destroy()

#第二題選擇（完成）
def btnnext():
    label.config(text="2014年世越號渡輪載著檀園高中的學生去旅行，卻不幸遇到離奇的船難\n當時船長下令所有乘客留在原地，如果你是當時的乘客，你會選擇......")
    l.config(text="你尚未選擇")
    for val,ch in c2.items():
        choice.append(tk.Radiobutton(window,text=ch,variable=var,value=val,command=select,font=('Arial', 20)))
        choice[-1].grid(row=val+3,rowspan=1,column=0,columnspan=5)
    label_img.destroy()
    btn.destroy()
    global bt2
    bt2.grid(row=7,column=4)

#第3題選擇（完成）
def btn3choice():
    label.config(text="(第三題內容)")
    l.config(text="你尚未選擇")
    for val,ch in c3.items():
        choice.append(tk.Radiobutton(window,text=ch,variable=var,value=val,command=select,font=('Arial', 20)))
        choice[-1].grid(row=val+3,rowspan=1,column=0,columnspan=5)
    label_img.destroy()
    btn2.destroy()
    global bt3
    bt3.grid(row=7,column=4)
    
#第4題選擇 （完成）
def btn4choice():
    label.config(text="(第4題內容)")
    l.config(text="你尚未選擇")
    for val,ch in c4.items():
        choice.append(tk.Radiobutton(window,text=ch,variable=var,value=val,command=select,font=('Arial', 20)))
        choice[-1].grid(row=val+4,rowspan=1,column=0,columnspan=5)
    label_img.destroy()
    btn3.destroy()
    global bt4
    bt4.grid(row=7,column=4)
    

#選擇的
def select():
    l.config(text=f"你選擇了{ca[j][var.get()]}")

global btstart
btstart=tk.Button(window,text="Start",command=bts,font=('Arial', 20))
btstart.grid(row=1)
bt1=tk.Button(window,text="Next",command=bt1,font=('Arial', 20))
bt2=tk.Button(window,text="Next",command=bt2,font=('Arial', 20))
bt3=tk.Button(window,text="Next",command=bt3,font=('Arial', 20))
bt4=tk.Button(window,text="Next",command=bt4,font=('Arial', 20))

window.mainloop()
```
